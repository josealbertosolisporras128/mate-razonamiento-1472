#### problema 1
La suma de un número más 9 En este caso, buscamos un valor desconocido que, al sumarle 9, nos dé como resultado 21.Paso 1: Planteamiento y ResoluciónDefinir la variable: Sea $x$ el número desconocido.Traducir al lenguaje algebraico: "La suma de un número ($x$) más el número 9 es igual a 21".$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Resultado:$$x = 12$$

```python
# Problema 1: Encontrar el número desconocido
# resultado_total = 21
# numero_conocido = 9
resultado_total=int(input("ingrese el resultado total: "))
numero_conocido=int(input("ingrese el numero conocido: "))
# Despeje: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado es: {x}")
```

#### problema 2
Las edades de Pedro y AnaAquí relacionamos dos cantidades basadas en una condición: la edad de uno depende de la del otro.Paso 1: Planteamiento y ResoluciónDefinir las variables:Edad de Ana: $a$Edad de Pedro: $2a$ (porque es el doble que la de Ana).Traducir al lenguaje algebraico: "Entre ambos suman un total de 36 años".$$a + 2a = 36$$Simplificar la ecuación: Sumamos las variables similares.$$3a = 36$$Despejar la variable: El 3 que está multiplicando a la $a$ pasa dividiendo al 36.$$a = \frac{36}{3}$$Resultado:Ana ($a$): $12$ años.Pedro ($2a$): $2 \times 12 = 24$ años.

```python
# Problema 2: Edades de Pedro y Ana
# suma_edades = 36
suma_edades=int(input("ingrese la suma de las edades: "))
# La ecuación simplificada es 3a = 36, por lo tanto a = 36 / 3
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
