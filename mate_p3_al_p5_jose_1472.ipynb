#### problema 3 La resta y el resultadoEnunciado:
Un número al que se le resta 7 da como resultado 18.Paso 1: Análisis y PlanteamientoIdentificar la incógnita: El "número" que no conocemos lo representamos con $x$.Traducir la operación: "Se le resta 7" significa que a nuestra $x$ le quitamos esa cantidad: $x - 7$.Establecer la igualdad: "Da como resultado 18" nos indica dónde va el signo igual:$$x - 7 = 18$$Paso 2: Resolución (Despeje)Para dejar sola a la $x$, observamos que el 7 tiene un signo negativo (está restando).Lo movemos al otro lado del signo igual con la operación opuesta, que es la suma:$$x = 18 + 7$$Calculamos el valor final:$$x = 25$$

```python
# Problema 3: x - 7 = 18
# resultado = 18
# resta = 7
resultado=int(input("ingrese el resultado: "))
resta=int(input("ingrese la resta: "))
# Despeje: x = 18 + 7
x = resultado + resta
print(f"El número del Problema 3 es: {x}")
```

#### problema 4  El triple y la suma combinados
Enunciado: El triple de un número más 4 es igual a 31.Paso 1: Análisis y PlanteamientoIdentificar la incógnita: Llamaremos al número $y$.Traducir "el triple": Multiplicamos el número por 3, es decir: $3y$.Traducir el resto: Sumamos 4 e igualamos a 31:$$3y + 4 = 31$$Paso 2: Resolución (Despeje)Primero movemos lo que suma o resta: El 4 que está sumando pasa al otro lado restando:$$3y = 31 - 4$$$$3y = 27$$Segundo movemos lo que multiplica: El 3 que multiplica a la $y$ pasa al otro lado dividiendo:$$y = \frac{27}{3}$$Calculamos el valor final:$$y = 9$$

```python
# Problema 4: 3y + 4 = 31
# total = 31
# suma = 4
# multiplicador = 3
total=int(input("ingrese el total: "))
suma=int(input("ingrese la suma: "))
multiplicador=int(input("ingrese el multiplicador: "))
# Despeje: y = (31 - 4) / 3
y = (total - suma) / multiplicador
print(f"El número del Problema 4 es: {int(y)}")
```

#### problema 5: Geometría y Perímetro
Enunciado: Un terreno tiene un largo que mide 5 metros más que el ancho. Si el perímetro total es de 50 metros, ¿cuánto mide cada lado?Paso 1: Análisis y PlanteamientoDefinir variables:Ancho = $w$Largo = $w + 5$ (porque mide 5 metros más que el ancho).Fórmula del perímetro: El perímetro es la suma de todos los lados de un rectángulo (2 anchos y 2 largos):$$Perímetro = 2(ancho) + 2(largo)$$Sustituir con nuestros datos:$$50 = 2(w) + 2(w + 5)$$Paso 2: Resolución (Despeje)Eliminar paréntesis: Multiplicamos el 2 por lo que está dentro:$$50 = 2w + 2w + 10$$Agrupar términos semejantes: Sumamos las $w$:$$50 = 4w + 10$$Despejar $w$:Pasamos el 10 restando: $50 - 10 = 4w$ $\Rightarrow$ $40 = 4w$.Pasamos el 4 dividiendo: $w = \frac{40}{4}$ $\Rightarrow$ $w = 10$.Encontrar dimensiones:Ancho: 10 metros.Largo: $10 + 5 = 15$ metros.

```python
# Problema 5: 2(w + 5) + 2w = 50 -> 4w + 10 = 50
# perimetro = 50
# diferencia_largo = 5
perimetro=int(input("ingrese el perimetro: "))
diferencia_largo=int(input("ingrese la diferencia de largo: "))
# Despeje: w = (50 - (2 * 5)) / 4
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Ancho del terreno: {ancho} m")
print(f"Largo del terreno: {largo} m")
```
